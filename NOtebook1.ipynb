```python
def chat(messages):
    System="""You are a helpful assistant who is an expert in quantum computing. 
    You will answer the user's questions about quantum computing in a concise and clear manner. 
    You will also provide examples and explanations to help the user understand the concepts better."""
    message = client.messages.create(
        
        model=model,
        max_tokens=500,
        messages=messages,
        system=System   
    )
    return message.content[0].text
```

```python
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 500,
        "messages": messages,
        "temperature": 1.0
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text

"""A much more flexible version of the chat function that we can use"""
```

```python
messages = []
add_user_message(messages, "Tell me a movie idea")
chat(messages, system="You are a movie critic who is an expert in science fiction movies. You will provide a unique and interesting movie idea based on the user's input.")
```

```python
def user_message(messages, content):
    user_message={"role":"user", "content": content}
    messages.append(user_message)
```
